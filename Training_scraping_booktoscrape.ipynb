```python
import requests
from bs4 import BeautifulSoup

url = "https://books.toscrape.com/"

response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")
category_container = soup.find("ul", class_="nav nav-list").find("ul").find_all("li")
for category in category_container:
    book = []
    category_name = category.find("a").text.strip()
    category_link = url + category.find("a")["href"]
    category_response = requests.get(category_link)
    category_soup = BeautifulSoup(category_response.text, "html.parser")
    book_container = category_soup.find("ol", class_="row").find_all("li")
    for book_title in book_container:
        book_name = book_title.find("img").get("alt")
        book.append(book_name)
    next_button = category_soup.find("li", class_="next")
    while next_button:
        next_page_link = (category_link + next_button.find("a")["href"]).replace("index.html", "")
        page2_response = requests.get(next_page_link)
        page2_soup = BeautifulSoup(page2_response.text, "html.parser")
        book_container2 = page2_soup.find("ol", class_="row").find_all("li")
        for book_title in book_container2:
            book_name = book_title.find("img").get("alt")
            book.append(book_name)
        next_button = page2_soup.find("li", class_="next")
    print(f"{category_name}:")
    for title in book:
        print(f" - {title}")


```
